```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
net = pd.read_csv('./data/sonoma-data-net.csv', parse_dates=['result_time'])
```

```python
sns.heatmap(net.groupby([net[' nodeid'], net[' epoch']]).size().unstack(0).sort_index(), cmap=sns.color_palette("crest", as_cmap=True))
```

```python
sns.heatmap(net.groupby([net[' nodeid'], net.result_time.dt.date]).size().unstack(0).sort_index(), cmap=sns.color_palette("crest", as_cmap=True))
```

```python
log = pd.read_csv('./data/sonoma-data-log.csv', parse_dates=['result_time'])
for col in log.columns:
    if col != col.strip():
        log.loc[:, col.strip()] = log[col]
        log.drop(col, axis=1, inplace=True)
```

```python
log.groupby('result_time').size()
```

```python
sns.heatmap(log.groupby([log.nodeid, log.epoch]).size().unstack(0).sort_index(), cmap=sns.color_palette("crest", as_cmap=True))
```
