```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
net = pd.read_csv('./data/sonoma-data-net.csv', parse_dates=['result_time'])
log = pd.read_csv('./data/sonoma-data-log.csv', parse_dates=['result_time'])
all = pd.read_csv('./data/sonoma-data-all.csv', parse_dates=['result_time'])

# clear excess white spaces in column names
for df in [net, log, all]:
    df.columns = [col.strip() for col in df.columns]
```

```python
sns.kdeplot(x=all.humid_temp[(all.humid_temp<=30)&(all.humid_temp>=0)])
```

```python
sns.scatterplot(y=net.humid_adj[(net.humidity>=0-3.5)&(net.humidity<=100+3.5)], x=net.humidity[(net.humidity>=0-3.5)&(net.humidity<=100+3.5)])
```

```python
sns.kdeplot(net.humidity[(net.humidity>=0-3.5)&(net.humidity<=100+3.5)], label='original')
sns.kdeplot(net.humid_adj[(net.humidity>=0-3.5)&(net.humidity<=100+3.5)], label='adjusted')
plt.legend()
plt.show()
```

```python
sns.kdeplot(log.humidity[(log.humidity>=0-3.5)&(log.humidity<=100+3.5)])
```

```python
net.groupby([' epoch']).size().plot()
```
